```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['xbox ', 'ybox ', 'width ', 'height', 'onpix ', 'xbar ',
       'ybar ', 'x2bar', 'y2bar ', 'xybar ', 'x2ybar', 'xy2bar', 'xedge ',
       'xedgey', 'yedge ', 'yedgex']].values
Y = df[['letter']].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, confusion_matrix,accuracy_score
from sklearn import metrics

```

```python
knn = KNeighborsClassifier(7)
```

```python
knn.fit(X_train,Y_train)
```

```python
Y_pred = knn.predict(X_test)
```

```python
Y_test = Y_test.flatten()
Y_pred = Y_pred.flatten()
```

```python
df_knn = pd.DataFrame({'Y_test': Y_test , 'Y_pred': Y_pred}) 
df_knn
```

```python
accuracy_score(Y_pred,Y_test)
```

```python
print(classification_report(Y_pred,Y_test))
```
