```python
import pandas as pd # data processing, CSV file
#pd: allows to use pandas functions pd.KAZA 
import numpy as nm  #Generate matrices
import matplotlib.pyplot as mtp  # for data visualization purposes
from sklearn.metrics import accuracy_score, classification_report 
# accuracy_score: function computes  accuracy 
# classification_report: performance evaluation metric in machine learning. It is used to show the precision, recall, F1 Score, and support of your trained classification model.
from sklearn.preprocessing import OneHotEncoder,StandardScaler
# OneHotEncoder string values zai letter column kda 
# StandardScaler to scale your data. twsee3 el data 
```

```python
DataSSet = pd.read_csv('letter-recognition.csv') #function help in raeding cvs file b access el dataset y3ne
DataSSet.head() #function is used to get the first n rows.
```

```python
DataSSet.shape  #17 column 20001 row 
```

```python
# Importing ll data set
y = DataSSet.iloc[:,0].values
X = DataSSet.iloc[:,1:17].values
# Splitting the dataset into the Training set and Test set  

from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state =0) 
print(DataSSet.iloc[:,1:17].values)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
# rule el probability 
classifier.fit(x_train, y_train)
# perform the calculation on the feature values of input data and fit this calculation to the transformer.
```

```python
y_pred  =  classifier.predict(x_test)
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test,y_pred)
print("Accuracy:" , ac)
```

```python
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
Accuracy = round((accuracy_score(y_test, y_pred) * 100))
from sklearn.metrics import confusion_matrix,accuracy_score
print("Accuracy : " , Accuracy , "%")
```
